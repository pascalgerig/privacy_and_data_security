```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace
```

# Wrapper for simple Laplace Distribution Sampling

```python
class LaplaceDistribution:
    def __init__(self, b):
        self.laplace = laplace(0, b)
        
    def sample(self):
        return self.laplace.rvs()
```

# Read CSV and convert to Array

```python
def readCsv(attribute_index=4):
    data_file = open('ex05-fake-registrations.csv', 'rU')
    first = True
    result = []
    for line in data_file:
        if first:
            first = False
            continue

        line = line.strip()
        line = line.replace(' ', '')
        temp = line.split(',')
        result.append(temp[attribute_index])
    return np.array(result)
    
```

# Create Histogram from Array

```python
def createHistogram(content):
    result = {}
    isNumbers = False
    try:
        int(content[0])
        isNumbers = True
    except ValueError:
        pass
    if isNumbers:
        # Put entries in bins of size 10 in array representation
        resultArray = np.zeros(10)
        for entry in content:
            resultArray[int(int(entry) / 10)] += 1
        
        # convert array to dict
        first = 0
        last = 9
        for value in resultArray:
            result[str(first) + "-" + str(last)] = value
            first += 10
            last += 10
    else:
        for entry in content:
            if entry in result:
                result[entry] += 1
            else:
                result[entry] = 1
    return result
```

```python
def addNoise(histogram, b):
    l = LaplaceDistribution(b)
    for entry in histogram:
        histogram[entry] += l.sample()
```

# Plot Histograms

```python
def plot_histogram(epsilon, attribute_index):
    csv = readCsv(attribute_index)
    histogram = createHistogram(csv)
    if epsilon != -1:
        addNoise(histogram,2/epsilon)
    plt.figure(figsize=(15, 15))
    plt.bar(histogram.keys(), histogram.values(), 0.9)
    plt.xticks(rotation=90)
    plt.show()

```
